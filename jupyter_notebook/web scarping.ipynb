```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
#url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
# load dataset into Pandas DataFrame
df = pd.read_csv("iris.csv")
feature = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
#print(feature)
# separating features
x = df.loc[:,feature]
#print(x)
# separating target
y = df.loc[:,'species']
#print(y)
#Standardising features
x = StandardScaler().fit_transform(x)
#print(x)
pca = PCA(n_components=2)
#print(pca)
pct = pca.fit_transform(x)
#print(pct)
principal_df = pd.DataFrame(pct,columns=['pc1','pc2'])
#print(principal_df)
finaldf= pd.concat([principal_df,df[['species']]],axis=1)
#print(finaldf)
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
ax = sns.scatterplot(x="pc1", y="pc2", hue="species",data=finaldf)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
df=pd.read_csv("wine.csv")
#print(df)
X = df.iloc[:, 0:13].values
#print(X)
y = df.iloc[:, 13].values
#print(y)
s= StandardScaler().fit_transform(X)
#print(X)
pca = PCA(n_components=4)
pt = pca.fit_transform(s)
principal_df = pd.DataFrame(pt,columns=['pc1','pc2',"pc3","pc4"])
finaldf= pd.concat([principal_df,df[['Customer_Segment']]],axis=1)
#print(finaldf)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=11116)
classifier = RandomForestClassifier(max_depth=2, random_state=16)
classifier.fit(X_train, y_train)
# Predicting the Test set results
y_pred = classifier.predict(X_test)
print('Accuracy',accuracy_score(y_test, y_pred))
# separating features
x2 = finaldf.loc[:,['pc1',"pc2","pc3","pc4"]]
# separating target
y2 = finaldf.loc[:,'Customer_Segment']
Xtrain, Xtest, ytrain, ytest = train_test_split(x2, y2, test_size=0.3, random_state=11116)
classifierpc = RandomForestClassifier(max_depth=2, random_state=0)
classifierpc.fit(Xtrain, ytrain)
# Predicting the Test set results
ypred = classifierpc.predict(Xtest)
print('Accuracy',accuracy_score(ytest, ypred))
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
import pandas as pd
fh=open("Chat.txt","r",encoding="utf8")
text=fh.read()
sid_obj = SentimentIntensityAnalyzer()

# polarity_scores method of SentimentIntensityAnalyzer
# object gives a sentiment dictionary. 
# which contains pos, neg, neu, and compound scores. 
sentiment_dict = sid_obj.polarity_scores(text) 
print("Overall sentiment dictionary is : ", sentiment_dict) 
print("Chat history was rated as ", sentiment_dict['neg']*100, "% Negative") 
print("Chat history was rated as ", sentiment_dict['neu']*100, "% Neutral") 
print("Chat history was rated as ", sentiment_dict['pos']*100, "% Positive")
print(sentiment_dict)
df=pd.DataFrame(sentiment_dict,columns =['neg', 'neu', 'pos'],index=[1,2,3])
df.plot.bar()

#print("Chat history Overall Rated As",end="") 

    # decide sentiment as positive, negative and neutral 
if sentiment_dict['compound'] >= 0.05 :
    print("Positive") 
elif sentiment_dict['compound'] <= - 0.05 :
    print("Negative") 
else :
    print("Neutral") 
```
